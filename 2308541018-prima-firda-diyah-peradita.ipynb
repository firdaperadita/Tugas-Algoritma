```python
# Jika diketahui bahwa n adalah dua digit terakhir dari nim saya (18) maka akan ditentukan bilangan prima antara 1000n dan 1000(n+1) 

import math

def apakah_prima(agk_18):
    if agk_18 <= 1:
        return False
    if agk_18 <= 3:
        return True
    if agk_18 % 2 == 0 or agk_18 % 3 == 0: 
        return False
    i = 5
    while i * i <= agk_18:
        if agk_18 % i == 0 or agk_18 % (i + 2) == 0:
            return False
        i += 6
    return True

# Menentukan nilai dari 1000n dan 1000(n+1) jika diketahui n adalah 2 digit terakhir nim saya 
n_18 = 18
x_18 = 1000 * n_18
y_18 = 1000 * (n_18 + 1)

# Mencari bilangan prima berdasarkan syarat yang tadi dicari 
bilangan_prima_18 = []
for agk_18 in range(x_18, y_18):
    if apakah_prima(agk_18):
        bilangan_prima_18.append(agk_18)

print(f"bilangan prima yang ada pada rentang angka {x_18} dan {y_18} adalah:")
print(bilangan_prima_18)
```
