```python
# tugas algoritma dan stuktur data
# menentukan persamaan kuadrat jika diketahui akar akarnya
# menentukan grafik fungsi kuadrat
# menentukan titik puncak dan sumbu simetri 
# menentukan posisi titik puncak (dikiri sumbu y, pada sumbu y, dikanan sumbu y)
# bagaimana kondisi parabola memotong sumbu y (potong sumbu y positif, potong sumbu y dipusat (0.0), potong sumbu y
# macam macam akar persamaan kuadrat dilihat dari diskriminan
# penyeleseian persamaan kuadrat dapat mebggubakan rums abc
# menyusun persamaan kuadrat baru jika diketahui akar akarny
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

# tentukan nilai a,b,c
print("pertama tama mari kita tentukan nilai a, b dan c terlebih dahulu")
a_18 = float(input("nilai a adalah ")) 
b_18 = float(input("nilai b adalah "))
c_18 = float(input("nilai c adalah "))
print("jadi kita temukan persamaan kuadratnya yaitu ({}x**2+{}x+{})".format(a_18,b_18,c_18))

d_18 = b_18**2-4*a_18*c_18 #menghitung diskriminan dengan d_18 sebagai tanda
# kemudian menenukan akar akar diskriminan nya
if d_18 > 0:
     print("nahh persamaan ini memiliki dua akar ril yang berbeda")
elif d_18 == 0:
     print( "nahh persamaan ini memiliki satu akar ril maka akar yang akan dihasilkan kembar")
else:
     print("nahh persamaan ini memiliki dua akar imajiner.")
# hitung akar-akarnya
a1_18 = (-b_18 - cmath.sqrt(d_18)) / (2 * a_18)
a2_18 = (-b_18 + cmath.sqrt(d_18)) / (2 * a_18)
print(f"nilai dari akar akarnya adalah ({a1_18}, {a2_18})")

#titik puncak dan sumbu simetri
x_18 = -b_18 / (2*a_18)
y_18 = a_18*x_18**2 + b_18*x_18 + c_18
print(f"nilai titik puncak nya adalah ({x_18})")
print(f"nilai sumbu simetri nya adalah({y_18})")  

# menentukan posisi titik puncak dan sumbu simetri y 
if x_18 < 0:
    print("titik puncak nya berada dikiri sumbu y")
elif x_18==0:
    print("titik puncak nya tepat pada sumbu y.")
else:
    print("titik puncak berada dikanan sumbu y.")
#kondisi parabola memotong sumbu y     
if c_18 > 0:
    print("parabola memotong sumbu y secara positif")
elif c_18==0:
    print("parabola memotong sumbu y tepat pada titik (0,0).")
else:
    print("parabola memotong sumbu y secara negatif.")

# menyusun persamaan kuadrat baru
p_18 = 1
e_18 = -2 * (a1_18 + a2_18 )
r_18 = a1_18  * a1_18 
print("persamaan kadrat baru nya adalah ({}x**2+{}x+{})".format(p_18,e_18,r_18))

#gambar grafik
x18_values = np.linspace(-20, 20, 400)
y18_values = a_18 * x18_values**2 + b_18 * x18_values + c_18
plt.plot(x18_values, y18_values, label=f'{a_18}x^2 + {b_18}x + {c_18}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafik Persamaan Kuadrat dan kuadrat baru')
plt.plot("persamaan kuadrat")
plt.plot("persamaan kuadrat baru")
plt.plot(y_18,'ro', label =f"titik puncak:({x_18},{y_18})")
plt.axvline(x=x_18,color ='red',linestyle='--',label=f"Sumbu simetri:x={x_18}")
plt.axhline(0, color='blue', linewidth=0.5)
plt.axvline(0, color='blue', linewidth=0.5)
plt.grid(color='lime', linestyle='--', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cari_persamaan_kuadrat(titik):
    a1_18, b1_18 = titik[0]
    a2_18, b2_18 = titik[1]
    a3_18, b3_18 = titik[2]
    
    # lalu membuat bentuk matriks nya
    a_18 = np.array([
        [a1_18**2, a1_18, 1],  
        [a2_18**2, a2_18, 1],
        [a3_18**2, a3_18, 1]
    ])
    b_18= np.array([b1_18, b2_18, b3_18])
    koefisien = np.linalg.solve(a_18,b_18)
    return koefisien
def masukkan_titik():
    titik = []
    for i in range(3):
        x_18 = float(input(f"masukkan titik x{i+1}: "))
        y_18 = float(input(f"masukkan titik y{i+1}: "))
        titik.append((x_18, y_18))
    return titik
# kemudian cari koefisien dari persamaan kuadrat
titik_18 = masukkan_titik()
a_18, b_18, c_18 = cari_persamaan_kuadrat(titik_18)
print(f"jadii persamaan kuadratnya yaitu y = {a_18:.2f}x^2 + {b_18:.2f}x + {c_18:.2f}")
x18_values = np.linspace(min(titik_18)[0] - 1, max(titik_18)[0] + 1, 400)
y18_values = a_18 * x18_values**2 + b_18 * x18_values + c_18

# gambar fungsi kuadrat
plt.plot(x18_values, y18_values, label=f'y = {a_18:.2f}x^2 + {b_18:.2f}x + {c_18:.2f}')
x18_points, y18_points = zip(*titik_18)
plt.scatter(x18_points, y18_points, color='black', zorder=5, label='titik yang digunakan')
plt.xlabel('x')
plt.ylabel('y')
plt.title('persamaan Kuadrat')
plt.legend()
plt.grid(True)
plt.axhline(0, color='blue', linewidth=0.5)
plt.axvline(0, color='red', linewidth=0.5)
plt.show()
```
