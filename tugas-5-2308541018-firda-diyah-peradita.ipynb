```python
import numpy as np
def volume_balok_18(panjang,lebar,tinggi):
    volume = panjang*lebar*tinggi
    return volume
def volume_tabung_18(jari_jari,tinggi):
    volume = np.pi*(jari_jari**2)*tinggi
    return volume
jenis_ruang_18 = input("jenis ruangnya adalah")
if jenis_ruang_18.upper() == 'BALOK':
    panjang = float(input('panjangnya adalah '))
    lebar = float(input("lebarnya adalah "))
    tinggi = float(input('tingginya adalah'))
    volume = volume_balok_18(panjang,lebar,tinggi)
    print(f"volume dari {jenis_ruang_18} adalah {volume}")
elif jenis_ruang_18.upper() == 'TABUNG':
    jari_jari = float(input('panjang jari jarinya adalah '))
    tinggi = float(input('tingginya adalah'))
    volume = volume_tabung_18(jari_jari,tinggi)
    print(f"volume {jenis_ruang_18} adalah {volume}")
    
```
